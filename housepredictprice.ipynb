# data inspection


```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
print(df.describe())
```

# Histogram for price values

```python
# Histogram for price
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='price', bins=10, kde=True)
plt.title('Distribution of House Prices')
plt.show()


```

# histogram for sqft living

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='sqft_living', bins=10, kde=True)
plt.title('Distribution of Living Area (sqft)')
plt.show()
```

# price vs sqft living

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sqft_living', y='price')
plt.title('Price vs Living Area')
plt.show()
```

# price vs bedrooms

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='bedrooms', y='price')
plt.title('Price by Number of Bedrooms')
plt.show()
```

# skewness & kurtosis

```python
from scipy.stats import skew, kurtosis

numeric_cols = df.select_dtypes(include=['int64','float64']).columns
for col in numeric_cols:
    print(f"{col}: Skewness={skew(df[col]):.2f}, Kurtosis={kurtosis(df[col]):.2f}")
```

# correlation matrix

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Correlation with price
print(corr['price'].sort_values(ascending=False))
```
